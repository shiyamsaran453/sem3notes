```python
import networkx as nx
```

```python
import matplotlib.pyplot as mat
%matplotlib notebook
```

```python
graph=nx.DiGraph(info="mark details")
```

```python
graph.add_node("mark",pos=(0,0))
graph.add_node("cat",pos=(-4,6))
graph.add_node("student",pos=(4,6))
graph.add_node("car",pos=(0,8))
graph.add_node("soccer player",pos=(-4,-6))
graph.add_node("spinoff",pos=(-5,0))
graph.add_node("csu",pos=(0,-8))
graph.add_node("sportsclub",pos=(5,0))
graph.add_node("addidas",pos=(4,-6))
pos=nx.get_node_attributes(graph,"pos")

```

```python
graph.add_edge("mark","cat",weight="has a")
graph.add_edge("mark","student",weight="is a")
graph.add_edge("mark","car",weight="has a")
graph.add_edge("mark","soccer player",weight="is a")
graph.add_edge("mark","spinoff",weight="is a captain of")
graph.add_edge("mark","sportsclub",weight="owned a ")
graph.add_edge("mark","csu",weight=" played in")
graph.add_edge("mark","addidas",weight="has a")
weight=nx.get_edge_attributes(graph,"weight")

```

```python
mat.figure()
nx.draw_networkx(graph,pos,font_weight="bold",node_size=2000,font_size=7)
nx.draw_networkx_edge_labels(graph,pos,edge_labels=weight)
```
