```python
import networkx as nx
```

```python
import matplotlib.pyplot as mat
%matplotlib notebook
```

```python
graph_saran=nx.DiGraph(info="sandeep and naveen details")
```

# creating nodes

```python
graph_saran.add_node("pmist",pos=(-1,0))
graph_saran.add_node("sandeep",pos=(4,-5))
graph_saran.add_node("naveen",pos=(-6,5))
graph_saran.add_node("tcs",pos=(4,5))
graph_saran.add_node("hr team",pos=(0,5))
graph_saran.add_node("development",pos=(4,1))
graph_saran.add_node("cricket",pos=(-6,-5))
graph_saran.add_node("basketball",pos=(-6,-1))
pos=nx.get_node_attributes(graph_saran,'pos')
```

# creating edges

```python
graph_saran.add_edge("sandeep","pmist",weight="studied in")
graph_saran.add_edge("naveen","pmist",weight="studied in")
graph_saran.add_edge("sandeep","development",weight="worked as a")
graph_saran.add_edge("naveen","hr team",weight="worked as a ")
graph_saran.add_edge("hr team","tcs",weight="in")
graph_saran.add_edge("development","tcs",weight="in")
graph_saran.add_edge("sandeep","cricket",weight="plays")
graph_saran.add_edge("naveen","basketball",weight="plays")
weight=nx.get_edge_attributes(graph_saran,"weight")
```

```python
mat.figure()
nx.draw_networkx(graph_saran,pos,node_size=2000,font_weight="bold",font_size=7)[]
nx.draw_networkx_edge_labels(graph_saran,pos,edge_labels=weight)
```
